```python
# You can also look for outliers in CPK using boxplot
sns.boxplot(x='creatinine_phosphokinase', data=df)
plt.title('Boxplot of Creatinine Phosphokinase (CPK)')
plt.xlabel('CPK Level')
plt.show()
```

```python
# Visualize the distribution of each numeric feature
numeric_features = ['ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium', 'time']
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

```

```python
# Visualize the count of binary features using countplots
binary_features = ['diabetes', 'high_blood_pressure', 'sex', 'smoking']
for feature in binary_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=feature, data=df)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()

```

```python
# You can also explore relationships between variables using scatter plots, box plots, etc.
# For example, to visualize the relationship between Ejection Fraction and Time:
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ejection_fraction', y='time', data=df)
plt.title('Relationship between Ejection Fraction and Time')
plt.xlabel('ejection fraction')
plt.ylabel('time (days)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], kde=True)
plt.title('Distribution of Age among Heart Failure Patients')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
death_rate_by_age = df.groupby('age')['DEATH_EVENT'].mean().reset_index()

# Visualize the relationship between age and death rate
plt.figure(figsize=(8, 6))
sns.lineplot(x='age', y='DEATH_EVENT', data=death_rate_by_age)
plt.title('Death Rate Variation with Age')
plt.xlabel('age')
plt.ylabel('Death Rate')
plt.show()
```

```python
# Calculate the percentage of male and female patients
gender_percentage = df['sex'].value_counts(normalize=True) * 100

# Visualize the percentage of male and female patients
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_percentage.index, y=gender_percentage.values)
plt.title('Percentage of Male and Female Patients')
plt.xlabel('gender')
plt.ylabel('percentage')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.show()
```

```python
age_bins = [0, 30, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['0-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
df['Age_Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Calculate platelet count statistics for each age group
platelet_stats_by_age = df.groupby('Age_Group')['platelets'].describe()

# Visualize the variation of platelet count among different age groups
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='platelets', data=df, order=age_labels)
plt.title('Platelet Count Variation Among Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('platelet Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the Pearson correlation coefficient between Creatinine and Sodium levels
correlation = df['serum_creatinine'].corr(df['serum_sodium'])

```

```python
# Visualize the relationship using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df)
plt.title('Correlation between Creatinine and Sodium levels')
plt.xlabel('serum creatinine')
plt.ylabel('serum sodium')
plt.text(3, 140, f'Correlation coefficient: {correlation:.2f}', fontsize=12)
plt.show()
```

```python
# Filter the dataset to include only male and female patients
male_patients = df[df['sex'] == 1]  # Assuming 1 represents male
female_patients = df[df['sex'] == 0]  # Assuming 0 represents female

# Calculate the prevalence of high blood pressure for male and female patients
male_high_bp_prevalence = male_patients['high_blood_pressure'].mean()
female_high_bp_prevalence = female_patients['high_blood_pressure'].mean()

# Visualize the prevalence of high blood pressure between male and female patients
plt.figure(figsize=(8, 6))
sns.barplot(x=['Male', 'Female'], y=[male_high_bp_prevalence, female_high_bp_prevalence])
plt.title('Prevalence of High Blood Pressure between Male and Female Patients')
plt.xlabel('gender')
plt.ylabel('Prevalence of High Blood Pressure')
plt.show()
```
